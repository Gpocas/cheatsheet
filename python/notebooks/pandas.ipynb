# **Map()**;

## Descrição:

A função map é destinada para realizar transformações de dados de uma lista ou Series.

### Vantagens:
A função map aplica o principio de vetorização em lista ou Series aumentando assim a performance.

### Desvatagens:
Como dito anteriormente o objetivo do map é apenas para transformação de dados. Sendo assim não é possivel implementar uma estruturas de decisão nessa função





```python
df.b.map(lambda x: x + 1)
```

```python
# Criando função lambda para testes
format = lambda x: x + 1

df.b.map(format)
```

# **Apply()**;

## Descrição:
a função apply assim como seu nome da a entender aplica uma função a uma lista ou Series de dados.

### Vantagens:
Ao contrario do map, com o apply é possivel aplicar uma função para transformação de dados com estruturas de decisão.

### Desvatagens:
A função aplly não aplica vetorização assim como map, ou seja aplly não tem um performance tão boa assim como map


```python
df.b.apply(lambda x: x + 1)
```

```python
df.b.apply(format) # isso tambem funciona
```

# **Pandas df.at | df.iat**

## df.at()
Acesse um único valor para um par de **`rótulos`** de linha/coluna.

Semelhante a loc, pois ambos fornecem pesquisas baseadas em **`rótulos`**. Use at se você precisar obter ou definir apenas um único valor em um DataFrame ou Series.

## df.iat()
Acesse um único valor para um par de linha/coluna por **`posição inteira`**.

Semelhante a iloc, pois ambos fornecem pesquisas baseadas em **`números inteiros`**. Use iat se você precisar obter ou definir apenas um único valor em um DataFrame ou Series.


```python
df = pd.DataFrame([[0, 2, 3], [0, 4, 1], [10, 20, 30]],
                  index=[4, 5, 6], columns=['A', 'B', 'C'])
df
```

```python
df.at[4, 'B']
```

```python
df.iat[0, 1]
```

# **Numpy Where**

## Descrição:
A função numpy.where() retorna os índices dos elementos em uma matriz de entrada onde a condição dada é satisfeita.

Sintaxe: numpy.where (condição [, x, y])
Parâmetros:
condição: Quando True, retorna x, caso contrário, retorna y.
x, y: valores para escolher. x, y e a condição precisam ser transmitidos para alguma forma.

Retorna:
out: [ndarray ou tupla de ndarrays] Se xey forem especificados, a matriz de saída contém elementos de x onde a condição é True e elementos de y em outro lugar.

Se apenas a condição for fornecida, retorne a tupla condição.nonzero(), os índices onde a condição é Verdadeira.

```python
import numpy as np

a = np.random.randn(2, 3)
print(a)
```

```python
b = np.where(a > 0, a, 0)

print(b)
```

```python
import numpy as np

a = np.arange(10)
b = np.where(a < 5, a, a * 10)

print(a)
print(b)
```

# **Filtros em DataFrames**

##Descrição:
Filtros em DataFrame como o proprio nome já diz, são formas na qual através de uma condição consulta\extrair dados em dataframes maiores.

```python
from pandas.io.formats.info import DataFrameTableBuilderNonVerbose
import pandas as pd

# assign data
df = pd.DataFrame({'Name': [' RACHEL  ', ' MONICA  ', ' PHOEBE  ',
                                   '  ROSS    ', 'CHANDLER', ' JOEY    '],

                          'Age': [30, 35, 37, 33, 34, 30],

                          'Salary': [100000, 93000, 88000, 120000, 94000, 95000],

                          'JOB': ['DESIGNER', 'CHEF', 'MASUS', 'PALENTOLOGY',
                                  'IT', 'ARTIST']})
```

```python
df
```

```python
df[df['Age'] > 30]
```

```python
df[(df['Age'] > 30) &
   (df['Salary'] > 93000) &
   (df['JOB'] == 'IT')]
```

```python
df[(df['Age'] > 30) &
   (df['Salary'] > 93000) &
   (df['JOB'] == 'IT')].Age
```

```python
df[(df['Age'] > 30) & (df['Salary'] == 94000)].Age[4] # '&' simbolo para condição 'and'
```

```python
df[(df['Age'] > 30) | (df['Salary'] == 94000)].Age[4]# '|' simbolo para condição 'or'
```

```python
df[(df['Age'] > 30) & (df['Salary'] == 94000)].Age.reset_index(drop=True)[0]
```

```python
df[df['JOB'].str.contains('(?i)(desi)', regex=True)]
```

```python
df[df['JOB'].str.match('(?i)(desi)')]
```
