# **Zip()**;

## Descrição:
A função zip é capaz de realizar um "join" entre duas ou mais tuplas/listas realizando a junção dos valores obrigatoriomente em um tupla

```python
a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica")

x = zip(a, b)

# use a função tuple() para exibir o valor do resultado:

print(tuple(x))
print(tuple(x)[0])
print(tuple(x)[0][0])

# ------------------------------------------------------------------------------
                      # Utilização da função zip com listas

a = ["John", "Charles", "Mike"]
b = ["Jenny", "Christy", "Monica"]

x = zip(a, b)
print(list(x))
print(list(x)[0])
print(list(x)[0][0])
```

# **Filter();**


## Descrição:
A filter função retorna um iterador onde os itens são filtrados por meio de uma função para testar se o item é aceito ou não.

```python
ages = [5, 12, 17, 18, 24, 32]

def myFunc(x):
  if x < 18:
    return False
  else:
    return True

adults = filter(myFunc, ages)

for x in adults:
  print(x)
```
