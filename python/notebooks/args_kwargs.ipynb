## **Utilização de Args na definição de Função:**

```python
# Exemplo da utilização de *args (sem diferença)

def sum_args(*args):
    print('e a soma é: ', sum(args))

def soma(lista):
    print('e a soma é: ', sum(lista))

sum_args(1,2,3)
soma([1,2,3])
```

```python
# Exemplo da utilização de *args (com diferença)

def sum_args(variavel, *args):
    print('minha variavel é: ', variavel)
    print('e a soma é: ', sum(args))

def soma(variavel, lista):
    print('minha variavel é: ', variavel)
    print('e a soma é: ', sum(lista))

sum_args(1,2,3)
soma([1,2,3])
```

## **Utilização de Kwargs na definição de Função:**

```python
# Exemplo da utilização de **kwargs (sem diferença)

def sum_kwargs(**kwargs):
    x = 0
    for key in kwargs.keys():
        x+=kwargs[key]
    print(x)
    
def soma_dict(dic):
    x = 0
    for key in dic.keys():
        x+=dic[key]
    print(x)

sum_kwargs(one=1, two=2, tree=3)
soma_dict({'one':1, 'two':2, 'tree':3})
```

```python
# Exemplo da utilização de **kwargs (sem diferença)

def sum_kwargs(number, **kwargs):
    x = number
    for key in kwargs.keys():
        x+=kwargs[key]
    print(x)
    
def soma_dict(dic):
    x = 0
    for key in dic.keys():
        x+=dic[key]
    print(x)

sum_kwargs(number=5, one=1, two=2, tree=3)
soma_dict({'number':5,'one':1, 'two':2, 'tree':3})
```

## **Utilização de Args e Kwargs na passagem de parametros para uma Função:**

```python
def teste(arg1, arg2, arg3, arg4):
    print(arg1)
    print(arg2)
    print(arg3)
    print(arg4)
    

args_list = [1,2,3,4]

args_dict = {
    "arg1": 1,
    "arg2": 2,
    "arg3": 3,
    "arg4": 4
}


teste(*args_list)
teste(**args_dict)
```

## **Utilização de Args no retorno de valores de uma Função:**

```python
def teste1():
    *lista, = [1,2,3,4,5]
    return *lista,


def teste2():
    lista = [*[1,2,3,4,5]]
    return lista


arg1, arg2, arg3, arg4, arg5 = teste1()
print(arg1)
print(arg2)
print(arg3)
print(arg4)
print(arg5)
```

## **Utilizando Args e Kwargs para realizar Merge de Listas e Dicionarios:**

```python
# merging_lists
my_first_list = [1, 2, 3]
my_second_list = [4, 5, 6]
my_merged_list = [*my_first_list, *my_second_list]

print(my_merged_list)
```

```python
# merging_dicts
my_first_dict = {"A": 1, "B": 2}
my_second_dict = {"C": 3, "D": 4}
my_merged_dict = {**my_first_dict, **my_second_dict}

print(my_merged_dict)
```

## **Utilização de Args no recebimento de valores de uma Função:**

```python
my_list = [1, 2, 3, 4, 5, 6]

a, *b, c = my_list

print(a)
print(b)
print(c)
```
