# List Comprehesion Classic 

Recebe uma lista e inclui a uma nova lista 50% do valor original 

```python
lista_precos = [500, 1000, 2000, 100, 25]

lista_com_imposto = [preco * 0.5 for preco in lista_precos]
print(lista_com_imposto)
```

# List Comprehesion With IF

Recebe uma lista e inclui a uma nova lista 50% do valor original apenas para os valores que são iguais ou maiores de 1000

```python
lista_precos = [500, 1000, 2000, 100, 25]

lista_com_imposto = [preco * 0.5  for preco in lista_precos if preco >= 1000]
print(lista_com_imposto)
```

# List Comprehesion With IF/ELSE

Recebe uma lista e inclui a uma nova lista 50% do valor original apenas para os valores que são iguais ou maiores de 1000, para os valores restante subtrai -50 do valor original 

```python
lista_precos = [500, 1000, 2000, 100, 25]

lista_com_imposto = [preco * 0.5 if preco >= 1000 else preco - 50 for preco in lista_precos]
print(lista_com_imposto)
```

# List Comprehesion With Operator AND | OR 

```python
new_list = [x for x in range(1, 10) if x % 2 and x % 3 == 0]
new_list = [x for x in range(1, 10) if x % 2 == 0 or x % 3 == 0]
```
