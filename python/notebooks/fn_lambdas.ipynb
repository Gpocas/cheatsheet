# Normal Function With Def

Exemplo de função utilizando def()

```python
preco = 500

def calcularImposto(preco):
    return preco * 0.3

print(calcularImposto(1000))
```

# Anonimal Function With Lambda

Exemplo de função anonima utilizando Lambda

```python
calcularImposto2 = lambda preco: preco * 0.3
calcularImposto2(100)
```

# Lambda Function With map

Exemplo de expressão lambda dentro de uma map()


```python
precos = [100, 500, 10, 25]

impostos = list(map(lambda x: x * 0.3, precos))
print(impostos)
```
