# Listas

1. Recomendados para estruturas homogêneas
2. Sintaxe usa []
3. Mutáveis
4. Mais lentas do que tuplas (ainda que a diferença só seja notável para muitos elementos)
5. Importante para manter ordem

```
lista_pares = [2, 4, 6, 8, 10]

lista_pares
```

```
lista_pares[0] = 3

lista_pares
```

```
lista_pares.append("segunda-feira")

lista_pares
```

# Tuplas

1. Sintaxe usa ()
2. Imutáveis
3. Mais rápidas do que listas (ainda que a diferença só seja notável para muitos elementos)
4. Ordem importa

```
dias_da_semana = ('segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sábado', 'domingo')

dias_da_semana[0] = "lucasfeira"
```

# Sets (Conjuntos)

1. Armazena elementos não ordenados e sem índices
2. Não permitem valores duplicados
3. São como os conjuntos da matemática
4. Rápidos

```
science_vlogs_set = set(['Canal do Pirula', 'Manual do Mundo', 'Nunca vi 1 Cientista', 'Canal do Pirula'])

science_vlogs_set.add('Universo Discreto')

science_vlogs_set
```

```
if 'Universo Discreto' in science_vlogs_set:
  print("está no grupo!")
else:
  print('não está')
```

```
valores = [1, 2, 3, 3, 4, 4, 5, 6, 6, 7, 1]

valores = list(set(valores))

valores
```

# Dicionários

1. Assim como seu análogo no mundo real, dicionários possuem entradas, associando elementos com chaves
2. Não mantém ordem
3. Rápidos

```
telefones = {'Lucas': '12345678', 'Fulano': '32134567'}

telefones['Lucas']
```

```
alunos = {20201234: ('Fulano', 20, "Ciência da Computação"), 20205678: ('Ciclana', 22, "Física")}

alunos
```

```
alunos = {20201234: ['Fulano', 20, "Ciência da Computação"], 20205678: ['Ciclana', 22, "Física"]}

alunos[20201234][0] = "Lucas"

alunos
```

## Quando usar **listas**?

Quando o conjunto pode ser alterado e a ordem importa. Lembre-se que duplicatas são permitidas. Listas longas são lentas.

## Quando usar **tuplas**?

Quando o conjunto não pode ser alterado. Lembre-se que a ordem importa e duplicatas são permitidas.

## Quando usar **sets**?

Precisa de uma estrutura pra saber se tem um elemento ou não? É o ideal.

É útil quando o conjunto não permite duplicatas. São bem mais rápidos do que listas, mesmo grandes.

## Quando usar **dicionários**?

Quando ao inserir entradas de dados, é preciso associar essas informações com uma chave e acessá-los de maneira rápida.
