## FindAll
retorna uma lista com todos os macths de uma string

````python
import re

txt = "onde posso pegar um onibus?"
x = re.findall("on", txt)
print(x)
````

## Split


## Positive LookBehind
O positive lookbehind em expressões regulares ((?<=...)) verifica se antes de uma posição específica na string existe um padrão definido. Ele não consome caracteres ou inclui o padrão no resultado da correspondência, apenas verifica sua existência.

#### Sintaxe e Estrutura:
```
(?<=pattern)match

```


#### Exemplo:

```python

import re

text = "Eu gosto de maçãs, mas não gosto de peras."
pattern = r"(?<!não\s)gosto"

matches = re.findall(pattern, text)
print(matches)  # Resultado: ['gosto']

```
---

## Negative LookBehind

O negative lookbehind em regex ((?<!...)) é usado para garantir que uma correspondência não é precedida por um padrão específico. Ele verifica o que vem antes de um ponto na string, sem incluir isso na correspondência final.

#### Sintaxe e Estrutura:
```
(?<!padrão_proibido)match"

```

#### Exemplo:
```python

import re

text = "Eu gosto de maçãs, mas não gosto de peras."
pattern = r"(?<!não\s)gosto"

matches = re.findall(pattern, text)
print(matches)  # Resultado: ['gosto']

```

No Python, há algumas regras importantes para usar negative lookbehinds:

O padrão dentro de (?<!...) deve ter largura fixa, como ocorre com o positive lookbehind. Ele só verifica a ausência de um padrão antes da posição atual e não consome caracteres.

---

## Positive Lookhead
O positive lookahead ((?=...)) verifica se um padrão específico está imediatamente após o ponto atual na string. Se o padrão for encontrado, a correspondência é considerada válida e continua; caso contrário, a correspondência falha

#### Sintaxe e Estrutura:
```
match(?=pattern)
```


#### Exemplo:
```python
import re

text = "Olá! Tudo bem! Vamos começar."
pattern = r"\w+(?=!)"

matches = re.findall(pattern, text)
print(matches)  # Resultado: ['Olá', 'bem']

```

---

## Negative LookHead
O Negative Lookahead é um operador em expressões regulares que permite verificar se um padrão específico não ocorre imediatamente após a posição atual na string. Assim como o lookahead positivo, ele não consome caracteres ou inclui o padrão na correspondência, apenas realiza uma verificação.

#### Sintaxe do Negative Lookahead:

```
match(?!pattern)

```

#### Exemplo:

```python
import re

text = "Olá! Tudo bem? Vamos começar."
pattern = r"\w+(?!\!)"

matches = re.findall(pattern, text)
print(matches)  # Resultado: ['Tudo', 'bem', 'Vamos', 'começar']

```
